```python
import pandas as pd
df = pd.read_csv(r'/content/DDH Data with Properties.csv')
```

```python
def bioactivity_class(std_value):
  if float(std_value) >= 10000:
    return "inactive"
  elif float(std_value) <= 1000:
    return "active"
  else:
    return "intermediate"
```

```python
df
```

```python
df['bioactivity_class'] = df['IC50(nM)'].apply(bioactivity_class)
```

```python
df
```

```python
df = df[df.bioactivity_class != 'intermediate']
```

```python
df
```

```python
import numpy as np
def pIC50(input):

  molar = input*(10**-9) # Converts nM to M
  ans = -np.log10(molar)

  return ans

def norm_value(input):
    if input > 100_000_000:
      return 100_000_000
    else:
      return(round(input))
```

```python
df['IC50(nM)_normed'] = df['IC50(nM)'].apply(norm_value)
```

```python
df['PIC50'] = df['IC50(nM)_normed'].apply(pIC50)
```

```python
def mannwhitney_modify(descriptors=[], verbose=False):
  # https://machinelearningmastery.com/nonparametric-statistical-significance-tests-in-python/
  from numpy.random import seed
  from numpy.random import randn
  from scipy.stats import mannwhitneyu

# seed the random number generator
  seed(1)
  final_df=pd.DataFrame()
# actives and inactives
  for desc in descriptors:
    selection = [desc, 'bioactivity_class']
    df = df_lipinski[selection]
    active = df[df.bioactivity_class == 'active']
    active = active[desc]

    selection = [desc, 'bioactivity_class']
    df = df_lipinski[selection]
    inactive = df[df.bioactivity_class == 'inactive']
    inactive = inactive[desc]

  # compare samples
    stat, p = mannwhitneyu(active, inactive)
    #print('Statistics=%.3f, p=%.3f' % (stat, p))

  # interpret
    alpha = 0.05
    if p > alpha:
      interpretation = 'Same distribution (fail to reject H0)'
    else:
      interpretation = 'Different distribution (reject H0)'

    results = pd.DataFrame({'Descriptor':desc,
                            'Statistics':stat,
                            'p':p,
                            'alpha':alpha,
                            'Interpretation':interpretation}, index=[0])

    final_df=pd.concat([final_df,results])

  return final_df.reset_index(drop=True)
```
